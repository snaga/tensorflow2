```python
from janome.tokenizer import Tokenizer
from gensim.models import word2vec
import zipfile
import os.path
import urllib.request as request
import re
```

```python
request.urlretrieve('http://www.aozora.gr.jp/cards/000081/files/462_ruby_716.zip', '462_ruby_716.zip')
z = zipfile.ZipFile('462_ruby_716.zip', 'r')
file = z.open('kazeno_matasaburo.txt', 'r')
bindata = file.read()
text_orig = bindata.decode('shift_jis')
```

```python
text = text_orig
text = re.split(r'\-{5,}', text)[2]
text = re.split(r'底本：', text)[0]
text = text.strip()
```

## 形態素解析

```python
t = Tokenizer()
```

```python
results = []
lines = text.split("\r\n")
for line in lines:
    s = line
    s = s.replace('|', '')
    s = re.sub(r'《.+?》', '', s)
    s = re.sub(r'［＃.+?］', '', s)
    tokens = t.tokenize(s)
    r = []
    for token in tokens:
        if token.base_form == "*":
            w = token.surface
        else:
            w = token.base_form
        ps = token.part_of_speech
        hinshi = ps.split(',')[0]
        if hinshi in ['名詞', '形容詞', '動詞', '記号']:
            r.append(w)
    rl = (" ".join(r)).strip()
    results.append(rl)
    print(rl)
```

## モデルファイルに書き出し

```python
wakachigaki_file = "matasaburo.wakachi"
with open(wakachigaki_file, "w", encoding='utf-8') as f:
    f.write("\n".join(results))
```

```python
data = word2vec.LineSentence(wakachigaki_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs=1, min_count=2, sg=1)
model.save('matasaburo.model')
```

```python
model.most_similar(positive=['学校'])
```
